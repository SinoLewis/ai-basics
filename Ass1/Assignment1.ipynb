```python
import pandas as pd
df = pd.read_csv('./Dataset.csv')
df.drop(['Ongata Rongai to Karen to Embakasi then back to Karen then Ongata Rongai'], axis="columns", inplace=True)
df.drop(df.index[[0]], inplace=True)
df.columns = df.iloc[0]
df.drop(df.index[[0]], inplace=True)
df.drop('Reg. No.', axis='columns', inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_time = LabelEncoder()
le_direction = LabelEncoder()
df['Direction'] = le_direction.fit_transform(df['Direction'])
df['Time '] = le_direction.fit_transform(df['Time '])
df
```

```python

df.columns
```

```python
from sklearn.model_selection import train_test_split

df.fillna(0)
x = df.drop(labels='Direction (Angle Changes)', axis=1)
y = df['Direction (Angle Changes)']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
X_train.tail()
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(X_train.fillna(0), y_train.fillna(0))

score = model.score(X_test.fillna(0), y_test.fillna(0))
print(f"Model Score is: {score}")
```

```python
prediction = model.predict([X_test.iloc[302]])
time = X_test['Time '].iloc[302]
print(f"A random prediction of {time} for Direction (Angle Changes) is {prediction}")
```
